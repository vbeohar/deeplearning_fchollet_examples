## Sample code on using PIL library and using it with Keras to preprocess the image into a 4D tensor

```python
# https://machinelearningmastery.com/how-to-load-convert-and-save-images-with-the-keras-api/
# https://github.com/fchollet/deep-learning-with-python-notebooks/blob/master/5.4-visualizing-what-convnets-learn.ipynb
```

```python
import keras
keras.__version__
```

```python
img_path = '/Users/Vaibhav_Beohar/Documents/VB_Mck_Docs/Other Trainings/deeplearning_fchollet_examples/data_samples/cat_sample.jpeg'
# We preprocess the image into a 4D tensor
from keras.preprocessing import image
import numpy as np
```

```python
img = image.load_img(img_path, target_size=(150, 150))
print(type(img))
print(img.format)
print(img.mode)
print(img.size)
img.show()
```

```python
# or show the image inline using matplotlib
%matplotlib inline
from matplotlib.pyplot import imshow
%matplotlib inline
imshow(np.asarray(img))
```

```python
img_tensor = image.img_to_array(img)
print(img_tensor.shape)
print(img_tensor)
```

```python
img_tensor = np.expand_dims(img_tensor, axis=0)
print(img_tensor.shape)
```

```python
# Remember that the model was trained on inputs
# that were preprocessed in the following way:
img_tensor /= 255.

# Its shape is (1, 150, 150, 3)
print(img_tensor.shape)
```
